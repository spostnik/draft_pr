### выгрузка данных по телеграмму и соц.сетям

```python
df_m_sn = pd.read_csv(r"..\data\График_для_www.pan-stp.com-Источники,_сводка(мессенджеры,соц.сети).csv")
df_m_sn
```

```python
df_m_sn.info()
```

```python
df_m_sn.columns = df_m_sn.columns.str.replace('Переходы из социальных сетей→', '')
df_m_sn
```

```python
df_m_sn.columns = df_m_sn.columns.str.replace('Переходы из мессенджеров→', '')
df_m_sn
```

```python
df_m_sn["Период"] = pd.to_datetime(df_m_sn["Период"])
```

```python
df_m_sn.info()
```

```python
df_m_sn.index = df_m_sn["Период"]
df_m_sn
```

```python
df_m_sn.drop(["Период"], axis=1, inplace=True)
df_m_sn
```

```python
df_teleg_soc = pd.DataFrame(df_m_sn)
df_teleg_soc
```
